In the dataset, certain variables are categorical and may contain missing values represented as zeros. Specifically, the variable "gender" has five missing values, while "self-employed" has 21 missing values.

```python
#Replace 0 in gender with the mode
gender_mode = loan_application[loan_application['Gender'] != '0']['Gender'].mode()[0]
loan_application['Gender'] = loan_application['Gender'].replace('0', gender_mode)
```

```python
#Replace N/As in Dependents with 0
loan_application['Dependents'] = loan_application['Dependents'].fillna(0)
```

```python
#Replace 0 in Self_Employed with No
loan_application['Self_Employed'] = loan_application['Self_Employed'].replace('0', 'No')
```

```python
#Replace N/As in Credit_History with 0
loan_application['Credit_History'] = loan_application['Credit_History'].fillna(0)
```

```python
loan_application.shape
```

The zeros in the "gender" variable are replaced with the mode. It is also assumed that zeros in the "self-employed" variable represent individuals who are not self-employed. Missing values in the "dependents" variable are filled with zero, based on the assumption that these indicate the absence of dependents. Similarly, missing values in the "credit history" variable are replaced with a designation indicating no credit history. After these adjustments, the shape of the dataset remains unchanged, indicating that no records were removed.

```python
#Replacing the outliers format to numerical
loan_application['CoapplicantIncome'] = loan_application['CoapplicantIncome'].replace('9.857.999.877.999.990', 9857999877999990)
loan_application['CoapplicantIncome'] = loan_application['CoapplicantIncome'].replace('1.612.000.084', 1612000084)
```

```python
#Converting CoapplicantIncome variable into float
loan_application['CoapplicantIncome']=loan_application['CoapplicantIncome'].astype('float')
```

```python
#Replace outlier in Coapplicantincome with the mean without the outliers
loan_application_without_outlier = loan_application[~loan_application['CoapplicantIncome'].isin([9857999877999990,1612000084])]
mean_coapplicant_income = loan_application_without_outlier['CoapplicantIncome'].mean()
loan_application['CoapplicantIncome'] = loan_application['CoapplicantIncome'].replace(9857999877999990, mean_coapplicant_income)
loan_application['CoapplicantIncome'] = loan_application['CoapplicantIncome'].replace(1612000084, mean_coapplicant_income)
```

The data contained some outliers in the "coapplicantIncome" variable. Initially, the values were converted to a numerical format and then to a float type. Subsequently, the outlier values were excluded, allowing for the calculation of the mean without the influence of these outliers. The calculated mean was then used to replace the outlier values.

### Data Split

```python
X = loan_application_mod.drop(columns=['Loan_Status','CoapplicantIncome'])
y = loan_application_mod['Loan_Status']

train_X, valid_X, train_y, valid_y = train_test_split(X,y, test_size=0.2, random_state=1, stratify=y)
```

Also for the data splitting it is important to get the best ratio for the modeling. In this case a 80% training and 20% testing is selected because as the data set is small, using too much for testing can reduce the training data which can hurt the model performance. An 80% training provides enough data for the model to learn meaningful patterns while also a 20% for testing is enough to get a reasonable estimate of performance on unseen data.

###Average Squared Error Tree (ASE)

```python
ASEClassTree = DecisionTreeRegressor(random_state=1)
ASEClassTree.fit(train_X,train_y)
```

```python
classificationSummary(train_y,ASEClassTree.predict(train_X))
```

```python
classificationSummary(valid_y,ASEClassTree.predict(valid_X))
```

```python
plotDecisionTree(ASEClassTree, feature_names=train_X.columns)
```

```python
# prompt: how many splits have this tree? and terminal nodes?

print('Number of splits: ', ASEClassTree.tree_.node_count - ASEClassTree.tree_.n_leaves)
print('Number of terminal nodes: ', ASEClassTree.tree_.n_leaves)
```

```python
# prompt: Create a roc index for this model

import matplotlib.pyplot as plt
# ROC Curve and AUC
fpr, tpr, thresholds = roc_curve(valid_y, ASEClassTree.predict(valid_X))
roc_auc = auc(fpr, tpr)

# Plot ROC curve
plt.figure()
plt.plot(fpr, tpr, color='darkorange', lw=2, label='ROC curve (area = %0.2f)' % roc_auc)
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic')
plt.legend(loc="lower right")
plt.show()

print("AUC:", roc_auc)
```
